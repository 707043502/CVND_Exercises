# Color Threshold, Green Screen

### Import resources

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import numpy as np
import cv2

%matplotlib inline
```

### Read in and display the image

```python
# Read in the image
image = mpimg.imread('images/car_green_screen.jpg')

# Print out the image dimensions (height, width, and depth (color))
print('Image dimensions:', image.shape)
```

```python
# Display the image
plt.imshow(image)
```

### Define the color threshold

```python
## TODO: Define our color selection boundaries in RGB values
lower_green = np.array([0,0,0]) 
upper_green = np.array([255,200,255])
```

### Create a mask

```python
# Define the masked area
mask = cv2.inRange(image, lower_green, upper_green)

# Vizualize the mask
plt.imshow(mask, cmap='gray')
```

```python
# Mask the image to let the car show through
masked_image = np.copy(image)

masked_image[mask == 0] = [0, 0, 0]

# Display it!
plt.imshow(masked_image)
```

### Mask and add a background image

```python

background_image = mpimg.imread('images/sky.jpg')




background_image = background_image[0:450, 0:660]
print(background_image.shape)

background_image[mask!=0] = [0, 0,0]

car=image.copy()

combined = background_image + masked_image
plt.imshow(combined)


```
